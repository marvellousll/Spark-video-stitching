```
import os
import cv2
import sys
from PIL import Image
```

```
def crop(path, image, num):
    # print("writing image ", num)
    im = Image.open(image)
    imgwidth, imgheight = im.size
    # print(imgwidth, ' ', imgheight)

    width = imgwidth/6
    dest = path + "/frame" + str(num)
    if not os.path.exists(dest):
        print(dest)
        os.makedirs(dest)

    j = 0
    for i in range(0, 8):
        # print("At epoch: ", i, "; left side position: ", j, "; right side position: ", j+width)
        box = (j, 0, j+width, 722)
        print(j, j+width)
        a = im.crop(box)
        try:
            a.save(os.path.join(dest, "frame%s.png" % (i+1)))
        except:
            print('FAILED! Position: ', num)
            pass
        j += width/2 
        j += width/10
```

```

video = cv2.VideoCapture('/content/test1.mov')
framePath = '/content/test'
name = 'frame'

sec = 0
frameRate = 0.05 # 20 fps
count=1

def videoToFrames(sec):
    video.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    success,image = video.read()
    if success:
    	# directory = framePath + '/' + 'frame' + str(count)
    	if not os.path.exists(framePath):
        	os.makedirs(framePath)
    	cv2.imwrite(os.path.join(framePath, name+str(count)+".jpg"), image)
      
    return success


successCode = videoToFrames(sec)

while successCode:
    count += 1
    sec = round(sec + frameRate, 2)
    successCode = videoToFrames(sec)
```

```
files = [f for f in os.listdir(framePath) if f.endswith(".jpg")]
```

```
outputPath = '/content/full_set'

for i in range(len(files)):
    filename = framePath + '/' + files[i]
    crop(outputPath, filename, i+1)
    #print(filename)
    #img = cv2.imread(filename)
    #frame_array.append(img)
```
